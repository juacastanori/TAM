```python
# ─── BLOQUE 2: generar app.py con rutas locales a los videos ────────────────
dashboard_code = '''
import streamlit as st
import os

st.set_page_config(page_title="Proyecto RL en ViZDoom", layout="wide")
st.title("🎮 Proyecto Final - Aprendizaje por Refuerzo en ViZDoom")
st.subheader("Universidad Nacional de Colombia - TAM 2025")
st.markdown("---")

# Carpeta base donde están los videos (ajusta si cambias el path)
BASE_VID_DIR = "ML_Theory_Repo/proyect_Doom/Videos"

# Carpeta base donde están las imagenes (ajusta si cambias el path)
BASE_IMG_DIR = "ML_Theory_Repo/proyect_Doom/Imagenes"

def show_video_if_exists(path, width=640):
    if os.path.exists(path):
        st.video(path)
    else:
        st.error(f"No se encontró el video: {path}")

def show_image_if_exists(path, width=600):
    if os.path.exists(path):
        st.image(path, width=width)
    else:
        st.error(f"No se encontró la imagen: {path}")


# Tabs de la Sección 1
tabs = st.tabs(["1️⃣ Introducción al Proyecto", "2️⃣ Introducción a ViZDoom",  "3️⃣ Escenario Basic", "4️⃣ Escenario DefendTheLine", "5️⃣ Escenario Perzonalizado"])

# ─────────────────────────────────────────────────────────────────────────────
# 1️⃣ SUBSECCIÓN: Introducción al Proyecto
with tabs[0]:
    st.header("🎯 ¿De qué trata este proyecto?")
    st.markdown("""
    Este proyecto explora el **entrenamiento de agentes inteligentes mediante Aprendizaje por Refuerzo Profundo (DRL)** en entornos 3D de **ViZDoom**.
    El objetivo es que los agentes aprendan estrategias para **combatir, navegar y sobrevivir** únicamente a partir de imágenes visuales, sin supervisión externa.

    Se implementaron y compararon dos algoritmos clave:
    - 🧠 **DQN y Dueling DQN**: variantes del método Q-Learning con redes neuronales.
    - 🧠 **PPO (Proximal Policy Optimization)**: método de tipo actor-crítico robusto para entornos visuales.

    El trabajo se desarrolló sobre escenarios con distinta dificultad, desde entornos simples de combate directo hasta mapas laberínticos con múltiples enemigos.
    """)

    st.subheader("🎞️ Video Introductorio")
    intro_path = os.path.join(BASE_VID_DIR, "video_introduccion.mp4")
    show_video_if_exists(intro_path)

    st.markdown("""
    Los experimentos se ejecutaron en un entorno de cómputo basado en notebooks de Kaggle, con acceso a una GPU (GPU Tesla P100), lo que permitió realizar múltiples millones de pasos de entrenamiento. Todo el código fue desarrollado en Python 3, utilizando PyTorch como framework principal para definir y entrenar los modelos de redes neuronales, debido a su flexibilidad y su compatibilidad con CUDA (GPU).
    """)

# ─────────────────────────────────────────────────────────────────────────────
# 2️⃣ SUBSECCIÓN: Introducción a ViZDoom
with tabs[1]:
    st.header("👾 ¿Qué es ViZDoom?")
    st.markdown("""
    **ViZDoom** es una plataforma basada en el motor del videojuego clásico Doom, diseñada para entrenar y evaluar algoritmos de inteligencia artificial en entornos tridimensionales con visión en primera persona.

    Es ampliamente utilizada en investigación de Aprendizaje por Refuerzo debido a que:
    - Permite observaciones visuales realistas (frames).
    - Simula condiciones del mundo real: enemigos, recursos limitados, navegación.
    - Es compatible con frameworks como **Gym, PyTorch, Stable-Baselines3, etc.**
    ---
    """)

    st.subheader("🎥 Ejemplos de entornos ViZDoom con políticas aleatorias")
    col1, col2 = st.columns(2)

    with col1:
        st.markdown("**📍 VizdoomBasic-v0**")
        show_video_if_exists(os.path.join(BASE_VID_DIR, "VizdoomBasic-v0.mp4"))

        st.markdown("**📍 VizdoomCorridor-v0**")
        show_video_if_exists(os.path.join(BASE_VID_DIR, "VizdoomCorridor-v0.mp4"))

        st.markdown("**📍 VizdoomDefendCenter-v0**")
        show_video_if_exists(os.path.join(BASE_VID_DIR, "VizdoomDefendCenter-v0.mp4"))

    with col2:
        st.markdown("**📍 VizdoomDefendLine-v0**")
        show_video_if_exists(os.path.join(BASE_VID_DIR, "VizdoomDefendLine-v0.mp4"))

        st.markdown("**📍 VizdoomHealthGathering-v0**")
        show_video_if_exists(os.path.join(BASE_VID_DIR, "VizdoomHealthGathering-v0.mp4"))

        st.markdown("**📍 VizdoomMyWayHome-v0**")
        show_video_if_exists(os.path.join(BASE_VID_DIR, "VizdoomMyWayHome-v0.mp4"))

        st.markdown("**📍 VizdoomTakeCover-v0**")
        show_video_if_exists(os.path.join(BASE_VID_DIR, "VizdoomTakeCover-v0.mp4"))


# 3️⃣ SUBSECCIÓN: ViZDoomBasic-V0
with tabs[2]:
    st.header("🎯 Escenario básico")
    st.markdown("""
    El primer entorno utilizado para entrenar y evaluar al agente es el escenario predefinido VizdoomBasic-v0, uno de los más simples disponibles en la plataforma ViZDoom. El mapa consiste en una sala rectangular con paredes, piso y techo grises. El jugador aparece centrado en una de las paredes largas, mientras que un enemigo (un monstruo rojo de forma circular) aparece aleatoriamente sobre la pared opuesta. El agente puede realizar únicamente tres acciones: moverse a la izquierda, moverse a la derecha y disparar. Un solo disparo acertado es suficiente para eliminar al enemigo. El episodio termina cuando el enemigo es eliminado o se alcanza un límite de tiempo de 300 \textit{tics} (pasos).

    Para el entrenamiento se utilizó una red Dueling DQN (Dueling Deep Q-Network), implementada en PyTorch. Esta arquitectura permite descomponer la estimación del valor-acción $Q(s,a)$ en dos componentes: el valor del estado $V(s)$ y la ventaja relativa $A(s,a)$.
    """)
    st.subheader("Especificaciones del modelo:")
    st.markdown("""
    - 🧠 Tres capas convolucionales seguidas por dos ramas densas independientes (valor y ventaja).
    - 🧠 Entrada: pila de 4 frames preprocesados 4 x 84 x 84 (Frames convertidos a escala de grises, con resolucion 84 x 84, apilados de a 4)
    - 🧠 Salida: valores $Q$ para cada acción.
    """)
    st.subheader("Configuración experimental")
    st.markdown("""
    - 🧠 Plataforma de entrenamiento: Kaggle Notebooks con GPU NVIDIA Tesla P100.
    - 🧠 Librerías: PyTorch, OpenCV, VizDoom v1.2.4, Gymnasium v0.28.1.
    - 🧠 Exploración: estrategia $\epsilon$-greedy con decaimiento exponencial
    - 🧠 Tamaño del batch 32.
    - 🧠 Memoria de experiencia:} buffer circular de 50,000 transiciones.
    - 🧠 Optimizador: Adam, tasa de aprendizaje $1 \times 10^{-4}$.
    - 🧠 Factor de descuento $\gamma$:} 0.99.

    El entrenamiento se realizo en dos fases, los códigos son los siguientes:
    """)
    st.markdown("""
    Puedes acceder al código en los siguientes enlaces:

    [Fase 1](https://github.com/juacastanori/TAM/blob/main/ViZDoom_Project/ViZDoom_BasicV0_ph1.ipynb)

    [Fase 2](https://github.com/juacastanori/TAM/blob/main/ViZDoom_Project/ViZDoom_BasicV0_ph2.ipynb)

    Para ejecutar la fase 2, primero se debe ejecutar el notebook de la fase 1, y cargar el modelo obtenido en el notebook de la fase 2, cambiando la ruta del modelo cargado en la celda de código correspondiente.
    """)

    st.markdown("""
    Puedes acceder al video del modelo en youtube

    [Video](https://youtu.be/Xyk-PK5OxaQ)
    """)


    st.subheader("Curvas de recompensa fase 1 ")
    intro_path = os.path.join(BASE_IMG_DIR, "Curva de aprendizaje.jpeg")
    show_image_if_exists(intro_path)

    st.subheader("Curvas de recompensa fase 2 ")
    intro_path = os.path.join(BASE_IMG_DIR, "Curva_Aprendizaje_Continuacion.jpeg")
    show_image_if_exists(intro_path)

    st.subheader("🎞️ Video del modelo: ")
    intro_path = os.path.join(BASE_VID_DIR, "ViZDoom_basicV0.mp4")
    show_video_if_exists(intro_path)

# 4️⃣ SUBSECCIÓN: VizdoomDefendLine-v0
with tabs[3]:
    st.header("🎯 Escenario Defend the Line")
    st.markdown("""
    Este escenario corresponde a la versión conocida como “Defend The Line” en ViZDoom. Es un escenario defensivo donde el agente permanece estático y debe eliminar enemigos que aparecen a lo lejos, antes de que logren infligirle daño. Se trata de un entorno clásico de visión en primera persona con tareas de disparo rápido y priorización de objetivos
    El espacio de acciones es discreto y corresponde a lo que define ViZDoom al crear el entorno, con botones como disparar y girar a la izquierda y a la derecha
    Se realiza el mismo preprocesamiento que para el modelo básico, convirtiendo la imagen a escalade grises, resolución 84 x 84, con una pila de 4 frames
    La arquitectura implementada es una red neuronal convolucional (CNN) empleada dentro del marco de un Deep Q-Network (DQN).
    """)
    st.subheader("Especificaciones del modelo:")
    st.markdown("""
    - 🧠 Arquitectura tipo Double DQN
    - 🧠 Tres capas convolucionales, cada una seguida por una ReLu
    - 🧠 Entrada: pila de 4 frames preprocesados 4 x 84 x 84 (Frames convertidos a escala de grises, con resolucion 84 x 84, apilados de a 4)
    - 🧠 Un vector aplanado de tamaño 7 x 7 x 64 = 3136 se conecta a una capa densa con activación ReLU. Finalmente, se conecta a una capa de salida  que devuelve los valores $Q(s,a)$ para cada acción disponible.
    """)
    st.subheader("Configuración experimental")
    st.markdown("""
    - 🧠 Plataforma de entrenamiento: Kaggle Notebooks con GPU NVIDIA Tesla P100.
    - 🧠 Librerías: PyTorch, OpenCV, VizDoom v1.2.4, Gymnasium v0.28.1.
    - 🧠 Pasos de entrenamiento: 5.000.000
    - 🧠 Exploración: estrategia $\epsilon$-greedy con decaimiento exponencial
    - 🧠 Tamaño del batch 32.
    - 🧠 Función de pérdida Huber (Smooth L1) entre los valores $Q$ predichos y los valores objetivo.
    - 🧠 Memoria de experiencia:} buffer circular de 100,000 transiciones.
    - 🧠 Optimizador: Adam, tasa de aprendizaje $1 \times 10^{-4}$.
    - 🧠 Factor de descuento $\gamma$:} 0.99.
    """)
    st.markdown("""
    Puedes acceder al código en los siguientes enlaces:

    [Código](https://github.com/juacastanori/TAM/blob/main/ViZDoom_Project/ViZDoom_Defendtheline.ipynb)
    """)

    st.markdown("""
    Puedes acceder al video del modelo en youtube

    [Video](https://youtu.be/onLg1Ou6Nks)
    """)


    st.subheader("Curvas de recompensas")
    intro_path = os.path.join(BASE_IMG_DIR, "Recompensas_DefendLine.jpeg")
    show_image_if_exists(intro_path)


    st.subheader("🎞️ Video del modelo: ")
    intro_path = os.path.join(BASE_VID_DIR, "ViZDoom_DefendLine-v0_VF.mp4")
    show_video_if_exists(intro_path)

# 4️5 SUBSECCIÓN: Escenario Personalizado
with tabs[4]:
    st.header("🎯 Escenario Personalizado")
    st.markdown("""
    Para evaluar el rendimiento de un agente PPO entrenado desde cero, se utilizó un entorno personalizado. El escenario consiste en un mapa tipo deathmatch con múltiples salas interconectadas, armas y pickups. Para esta implementación, primeramente se eliminaron todos los enemigos del entorno y se enfocó únicamente en recompensar al agente por explorar y recolectar objetos.

    El entorno fue implementado como un wrapper personalizado compatible con Gym, con observaciones de tipo imagen RGB de tamaño 160 x 120, y recompensas diseñadas cuidadosamente (reward shaping). Las recompensas contemplan eventos como: cambio de arma, recolección y uso de munición, daño infligido, mejoras en salud y armadura, desplazamiento efectivo en el mapa, y enemigos eliminados, para ser reutilizado para hacer el mismo entrenamiento, pero ahora agregando enemigos.

    Para diseñar un entorno personalizado con técnicas de aprendizaje por refuerzo más efectivas, se consideraron tanto las señales del entorno (variables internas del juego) como un conjunto discreto de acciones posibles para el agente.

    """)
    st.subheader("Especificaciones del modelo:")
    st.markdown("""

    Para este proyecto se implementó un agente basado en el algoritmo Proximal Policy Optimization (PPO), una técnica de aprendizaje por refuerzo de tipo actor-crítico con políticas estocásticas y actualización confiable por clipping. PPO ha demostrado un desempeño robusto en entornos visuales complejos como ViZDoom debido a su balance entre estabilidad y eficiencia computacional. Se diseñó una red convolucional adaptada a las entradas visuales del entorno, que consisten en stacks de 4 imágenes RGB (resolución 160×120). Esta red reemplaza al extractor de características por defecto de Stable-Baselines3, y tiene la siguiente arquitectura:

    - 🧠 Arquitectura tipo PPO implementada con SB3
    - 🧠 Entrada: Tensor de forma (4, 120, 160), normalizado en el rango [0, 1].
    - 🧠 3 Capas convolucionales cada una seguida por una ReLu
    - 🧠 Una capa densa con 512 unidades + ReLU.

    Para acelerar el entrenamiento y reducir la varianza del gradiente, se empleó un esquema de entornos paralelos con 8 entornos simultáneos.
    """)
    st.subheader("Configuración experimental")
    st.markdown("""
    - 🧠 Plataforma de entrenamiento: Kaggle Notebooks con GPU NVIDIA Tesla P100.
    - 🧠 Librerías: PyTorch, Stable-Baselines3, OpenCV, VizDoom v1.2.4, Gymnasium v0.28.1.
    - 🧠 Pasos de entrenamiento: 10.000.000
    - 🧠 Exploración: estrategia $\epsilon$-greedy con decaimiento exponencial
    - 🧠 Tamaño del batch 1024, con 4 epocas.
    - 🧠 Función de pérdida Huber (Smooth L1) entre los valores $Q$ predichos y los valores objetivo.
    - 🧠 Memoria de experiencia:} buffer circular de 100,000 transiciones.
    - 🧠 Optimizador: Adam, tasa de aprendizaje $1 \times 10^{-4}$.
    - 🧠 Factor de descuento $\gamma$:} 0.99.
    """)
    st.markdown("""
    Puedes acceder al código en los siguientes enlaces:

    [Código Entrenamiento](https://github.com/juacastanori/TAM/blob/main/ViZDoom_Project/ViZDoom_Custom_map_nobot.ipynb)

    Para observar el video, subir el modelo obtenido, y cambiar la ruta del modelo subido en el siguiente notebook:

    [Código Evaluación](https://github.com/juacastanori/TAM/blob/main/ViZDoom_Project/ViZDoom_Custom_nobot_video.ipynb)
    """)

    st.markdown("""
    Puedes acceder al video del modelo en youtube

    [Video](https://youtu.be/wakcjBQBmAU)
    """)


    st.subheader("Curvas de recompensas")
    intro_path = os.path.join(BASE_IMG_DIR, "nobots_reward.jpeg")
    show_image_if_exists(intro_path)


    st.subheader("🎞️ Video del modelo: ")
    intro_path = os.path.join(BASE_VID_DIR, "ViZDoom_Custom_Map_Nobot.mp4")
    show_video_if_exists(intro_path)

'''



with open("app.py", "w", encoding="utf-8") as f:
    f.write(dashboard_code)
```

```python
# ─── BLOQUE 3: Ejecutar Streamlit y abrir túnel Cloudflare ────────────────
import os
import threading
import time

def run_streamlit():
    os.system("streamlit run app.py --server.port 8501")

def launch_app():
    thread = threading.Thread(target=run_streamlit)
    thread.start()
    time.sleep(5)  # dar tiempo a que arranque streamlit

    # Descargar cloudflared y ejecutar
    !wget -q https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64 -O cloudflared
    !chmod +x cloudflared
    print("🔁 Lanzando túnel Cloudflare...")
    !./cloudflared tunnel --url http://localhost:8501

launch_app()
```
