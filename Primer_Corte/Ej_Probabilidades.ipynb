Prompt:Estoy realizando un ejercicio de probabilidades basado en el teorema de bayes en python. Tengo 2 cajas, una roja y otra azul, y en cada caja hay cierta cantidad de manzanas y naranjas, 6 naranjas y 2 manzanas en la caja roja, y 3 manzanas y una naranja en la caja azul. Tengo las probabilidades marginales P(B=r)=0.4; y P(B=b)=0.6 (marginales de caja roja o azul). Tengo que encontrar la probabilidad de que la fruta sea manzana o naranja y encontrar las 4 probabilidades condicionales de, dado escoger una caja se obtiene tal fruta P(B|F).

```python
# Probabilidades marginales de las cajas
P_B_r = 0.4
P_B_b = 0.6

# Probabilidades condicionales de cada fruta dado una caja
P_F_m_B_r = 2/8  # Probabilidad de manzana dado caja roja
P_F_n_B_r = 6/8  # Probabilidad de naranja dado caja roja
P_F_m_B_b = 3/4  # Probabilidad de manzana dado caja azul
P_F_n_B_b = 1/4  # Probabilidad de naranja dado caja azul

# Probabilidades totales de las frutas
P_F_m = P_F_m_B_r * P_B_r + P_F_m_B_b * P_B_b
P_F_n = P_F_n_B_r * P_B_r + P_F_n_B_b * P_B_b

# Aplicamos el teorema de Bayes para calcular P(B|F)
P_B_r_F_m = (P_F_m_B_r * P_B_r) / P_F_m
P_B_b_F_m = (P_F_m_B_b * P_B_b) / P_F_m
P_B_r_F_n = (P_F_n_B_r * P_B_r) / P_F_n
P_B_b_F_n = (P_F_n_B_b * P_B_b) / P_F_n

# Imprimimos los resultados
print(f"P(F=m): {P_F_m:.4f}")
print(f"P(F=n): {P_F_n:.4f}")
print(f"P(B=r|F=m): {P_B_r_F_m:.4f}")
print(f"P(B=b|F=m): {P_B_b_F_m:.4f}")
print(f"P(B=r|F=n): {P_B_r_F_n:.4f}")
print(f"P(B=b|F=n): {P_B_b_F_n:.4f}")
```

```python
# Entrada de datos por el usuario
num_cajas = int(input("Ingrese el número de cajas (1-3): "))
num_objetos = int(input("Ingrese el número de tipos de objetos en las cajas (1-3): "))

cajas = {}
prob_cajas = {}
objetos = []

# Ingreso de nombres de objetos
definir_objetos = True
for i in range(num_objetos):
    objeto = input(f"Ingrese el nombre del objeto {i+1}: ")
    objetos.append(objeto)

# Ingreso de datos por caja
for i in range(num_cajas):
    nombre_caja = input(f"Ingrese el nombre de la caja {i+1}: ")
    cajas[nombre_caja] = {}
    prob_cajas[nombre_caja] = float(input(f"Ingrese la probabilidad marginal de elegir la caja {nombre_caja}: "))
    for objeto in objetos:
        cajas[nombre_caja][objeto] = int(input(f"Ingrese la cantidad de {objeto} en la caja {nombre_caja}: "))

# Cálculo de probabilidades condicionales
prob_objetos = {objeto: 0 for objeto in objetos}
prob_condicionales = {objeto: {} for objeto in objetos}

for objeto in objetos:
    for caja in cajas:
        total_caja = sum(cajas[caja].values())
        prob_condicionales[objeto][caja] = cajas[caja][objeto] / total_caja
        prob_objetos[objeto] += prob_condicionales[objeto][caja] * prob_cajas[caja]

# Aplicación del teorema de Bayes
prob_cajas_dado_objeto = {objeto: {} for objeto in objetos}
for objeto in objetos:
    for caja in cajas:
        prob_cajas_dado_objeto[objeto][caja] = (prob_condicionales[objeto][caja] * prob_cajas[caja]) / prob_objetos[objeto]

# Imprimimos los resultados
for objeto in objetos:
    print(f"P(F={objeto}): {prob_objetos[objeto]:.4f}")
    for caja in cajas:
        print(f"P(B={caja}|F={objeto}): {prob_cajas_dado_objeto[objeto][caja]:.4f}")
```
